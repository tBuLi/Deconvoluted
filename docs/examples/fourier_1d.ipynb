# 1D Fourier transform


```python
import numpy as np
import matplotlib.pyplot as plt

from deconvoluted import fourier_transform
```

Suppose we want to compute the 1D fourier transform $F(\nu)$ of a 
function $f(t)$. Let us generate some data which has a frequency of $0.2$ Hz 
in the $x$ direction, and $0.1$ Hz in the $y$ direction:

```python
t = np.linspace(0, 20, 201)  # 20 seconds
nu_1 = 1
nu_2 = 3
f_t = np.sin(nu_1 * 2 * np.pi * t) + np.sin(nu_2 * 2 * np.pi * t)

plt.plot(t, f_t)
plt.xlabel(r'$t$ / s')
plt.ylabel(r'$f(t)$')
plt.show()
```

Taking the transform is now simply a matter of calling `fourier_transform`: 

```python
F_nu, nu = fourier_transform(f_t, t)
```

```python
plt.plot(nu, F_nu)
plt.xlabel(r'$\nu$ / $s^{-1}$')
plt.ylabel(r'$F(\nu)$')
plt.show()
```

As expected, we find resonances at $\nu_1 = 1$ Hz and $\nu_2 = 3$ Hz.
